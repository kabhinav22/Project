```python
## some basic visualization
```

```python
plt.figure(figsize=(12, 6))  # Set the figure size

plt.plot(df['High'], label='High Price', color='blue')



plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot for High')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Low'], label='Low Price', color='red')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot for Low')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Volume'], label='Volume', color='green')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot for High')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.plot(df['Open'], label='Open', color='orange')
plt.plot(df['Close'], label='Close', color='blue')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot for Open and Close')

plt.legend()

plt.show()
```
